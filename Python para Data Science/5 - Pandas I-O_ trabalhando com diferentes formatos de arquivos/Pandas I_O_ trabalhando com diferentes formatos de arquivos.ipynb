###Google Planilhas

```python
#https://docs.google.com/spreadsheets/d/1WBA00sG0tEkvL99qFbY4MHsQgeKPtYFfNefeUoHrfug/edit?usp=sharing
```

Toda planilha possui um ID para identificá-la. Este ID está presente nessa URL, entre   **/d/**    e   **/edit?usp=sharing.**

```python
sheet_id = '1WBA00sG0tEkvL99qFbY4MHsQgeKPtYFfNefeUoHrfug'
```

A seguir, adicionaremos alguns parâmetros para permitir o acesso à planilha. Vamos colocar uma barra após {sheet_id} e inserir o parâmetro **gviz/tq:**

O que é esse parâmetro e para que ele serve? Para ter acesso à planilha, é necessário acessar a API de visualização de dados do Google. API significa Application Programming Interface, isto é, uma interface de programação de aplicações. Elas permitem sistemas diferentes se comuniquem.

No caso, queremos criar a comunicação entre o Google Colab e o Google Planilhas. O parâmetro gviz/tq permite o acesso à API responsável por essa comunicação.

Após **gviz/tq**, adicionaremos um ponto de interrogação e o parâmetro **tqx=out:csv**:

```python
url = f'https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet'
```

```python
dados_co2_sheets = pd.read_csv(url)
```

```python
dados_co2_sheets.head()
```

Para especificar a página da planilha que queremos acessar, basta inserir um símbolo de igual ao final da URL, seguida da variável sheet_name entre chaves:

```python
sheet_id = '1WBA00sG0tEkvL99qFbY4MHsQgeKPtYFfNefeUoHrfug'
sheet_name = 'emissoes_percapita'
url_percapita = f'https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}'
```

```python
percapita_sheets = pd.read_csv(url_percapita)
```

```python
percapita_sheets.head()
```

```python
sheet_id = '1WBA00sG0tEkvL99qFbY4MHsQgeKPtYFfNefeUoHrfug'
sheet_name = 'fontes'
url_fontes = f'https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}'
```

##Preparando o Banco SQL

```python
from sqlalchemy import create_engine, MetaData, Table, inspect
```

```python
engine = create_engine('sqlite:///:memory:')
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'
```

```python
import pandas as pd
```

```python
dados = pd.read_csv(url)
```

```python
dados.head()
```

```python
dados.to_sql('Clientes', engine, index=False)
```

```python
inspector = inspect(engine)
```

```python
print(inspector.get_table_names())
```

## SELECT

```python
query = 'SELECT * FROM clientes WHERE Categoria_de_renda = "Empregado"'
```

```python
empregados = pd.read_sql(query, engine)
```

```python
empregados
```

```python
empregados.to_sql('empregados', con=engine, index=False)
```

```python
pd.read_sql_table('empregados', engine)
```

```python
pd.read_sql_table('empregados', engine, columns=['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual'])
```

A função **read_sql** da biblioteca Pandas do Python é utilizada para executar uma **consulta SQL** em um banco de dados e retornar os resultados em um objeto DataFrame.

Já a função **read_sql_table** é usada para carregar dados diretamente de uma tabela existente no banco de dados e não para executar consultas personalizadas.

##DELETE

```python
query = 'SELECT * FROM clientes'
```

```python
pd.read_sql(query, engine)
```

```python
query = 'DELETE FROM clientes WHERE ID_Cliente =5008804'
with engine.connect() as conn:
  conn.execute(query)
```

```python
pd.read_sql_table('clientes', engine)
```

##UPDATE

```python
query = 'UPDATE clientes SET Grau_escolaridade="Ensino Superior" WHERE ID_cliente=5008808'
with engine.connect() as conn:
  conn.execute(query)
```

```python
pd.read_sql_table('clientes', engine)
```
