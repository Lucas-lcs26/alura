# Manipulando os dados

## Criando colunas numéricas

```python
url = "https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv"
dados = pd.read_csv(url, sep =';')
dados.head()
```

```python
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
```

```python
dados.head()
```

```python
dados['Valor_por_ano'] = dados['Valor']*12 + dados['IPTU']
```

```python
dados.head()
```

## Criando colunas categóricas

```python
dados['Descriao'] = dados['Tipo'] + ' em ' +dados['Bairro']
dados.head()
```

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + dados['Quartos'].astype(str) + ' quarto(s) e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
dados.head()
```

```python
dados['Descricao']
```

```python
dados.drop(['Descriao'], axis=1, inplace= True)
dados
```

```python
dados['Possui_Suit'] = dados['Suites' ].apply(lambda x: "Sim" if x > 0 else "Não")
dados.head()
```

```python
dados.to_csv('Dados_completos_Dev.csv', index=False, sep= ';')
```
