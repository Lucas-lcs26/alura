```python
link = 'https://raw.githubusercontent.com/allanspadini/numpy/dados/citrus.csv'
citrus = np.loadtxt(link,delimiter=',',usecols=np.arange(1,6,1),skiprows=1)
```

```python
citrus.shape
```

```python
citrus
```

```python
diametro_laranjas = citrus[:5000,0]
peso_laranjas = citrus[:5000,1]
diametro_toranjas = citrus[5000:,0]
peso_toranjas = citrus[5000:,1]
```

```python
import matplotlib.pyplot as plt
plt.plot(diametro_laranjas,peso_laranjas)
plt.plot(diametro_toranjas, peso_toranjas)
```

```python
Y = peso_laranjas
X = diametro_laranjas
n = np.size(X)
```

```python
a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y)) / (n*np.sum(X**2)-np.sum(X)**2)
a
```

```python
b = np.mean(Y) - a*np.mean(X)
b
```

```python
y = a * X + b
plt.plot(diametro_laranjas,y)
plt.plot(diametro_laranjas,peso_laranjas)
```

```python
np.linalg.norm(peso_laranjas-y)
```
