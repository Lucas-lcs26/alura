Chegou o momento de praticar! Vamos aplicar os conceitos aprendidos durante a aula a partir de algumas atividades. Solucione os problemas propostos através de códigos utilizando a base de dados disponibilizada no curso.

1) Faça um agrupamento de dados com base na coluna "Nível 1 - Setor" para visualizar o dicionário contendo as chaves de grupos formados e a lista de índices de cada grupo.

2) Faça um agrupamento de dados com base na coluna "Nível 1 - Setor" e localize os dados do grupo "Agropecuária".

3) Faça um agrupamento de dados com base na coluna "Nível 1 - Setor" para identificar a média de emissão de cada atividade econômica no ano de 2021.

4) Faça um agrupamento de dados com base na coluna "Nível 1 - Setor" para identificar a soma de emissão de cada atividade econômica. Ordene os dados da maior para menor emissão.

Caso precise de ajuda, opções de solução das atividades estarão disponíveis na seção “Opinião da pessoa instrutora”.

```python
#primeiro estou deixando as tabelas com o mesmo formato das aulas, usando o metodo melt() para colocar os anos em linhas:

emissoes_gases = emissoes_gases[emissoes_gases['Emissão / Remoção / Bunker'] == 'Emissão']
emissoes_gases = emissoes_gases.drop(columns = 'Emissão / Remoção / Bunker')
colunas_info = list(emissoes_gases.loc[:,'Nível 1 - Setor':'Produto'].columns)
colunas_emissão = list(emissoes_gases.loc[:,1970:2021].columns)
emissao_por_ano = emissoes_gases.melt(id_vars= colunas_info, value_vars= colunas_emissão, var_name= 'Ano', value_name= 'Emissão')
emissao_por_ano
```

```python
#1) Faça um agrupamento de dados com base na coluna "Nível 1 - Setor" para visualizar o dicionário contendo as chaves de grupos formados e a lista de índices de cada grupo.
emissao_por_ano.groupby('Nível 1 - Setor').groups
```

```python
#2) Faça um agrupamento de dados com base na coluna "Nível 1 - Setor" e localize os dados do grupo "Agropecuária".

emissao_por_ano.groupby('Nível 1 - Setor').get_group('Agropecuária')
```

```python
#3) Faça um agrupamento de dados com base na coluna "Nível 1 - Setor" para identificar a média de emissão de cada atividade econômica no ano de 2021.

emissao_por_ano[emissao_por_ano['Ano'] == 2021].groupby('Nível 1 - Setor')[['Emissão']].mean()
```

```python
#4) Faça um agrupamento de dados com base na coluna "Nível 1 - Setor" para identificar a soma de emissão de cada atividade econômica. Ordene os dados da maior para menor emissão.

emissao_por_ano.groupby('Nível 1 - Setor')[['Emissão']].sum().sort_values('Emissão', ascending= False)
```

###Desafio: emissão por setor econômico

Uma das tarefas passadas pela equipe de supervisão do órgão ambiental era de encontrar a emissão média de cada gás por ano. Conseguimos realizar essa tarefa usando o método groupby() e pivot_table() para gerar uma tabela contendo a emissão ao longo dos anos para cada um dos gases.

A equipe de supervisão deseja saber agora a média de emissão para cada um dos setores econômicos da coluna "Nível 1 - Setor". Como desafio, crie uma tabela de pivô e uma visualização gráfica contendo essas informações, utilizando apenas o método pivot_table para a construção da tabela, sem utilizar o método groupby.

Você pode buscar informações de como fazer isso na [documentação Pandas sobre o método pivot_table](https://pandas.pydata.org/docs/reference/api/pandas.pivot_table.html).

```python
emissao_por_ano.head(2)
```

```python
emissão_media_setores = emissao_por_ano.pivot_table(index = 'Ano', columns= 'Nível 1 - Setor', values= 'Emissão', aggfunc='mean')
emissão_media_setores
```

```python
emissão_media_setores.plot(subplots=True, figsize = (10,15));
```

```python
#1) Faça um agrupamento de dados com as colunas "Estado" e "Nível 1 - Setor", obtendo a soma de emissão e armazenando o resultado em uma tabela.

emissao_por_ano.groupby(['Estado', 'Nível 1 - Setor'])[['Emissão']].sum()
```

```python
#2) Utilizando a tabela construída na atividade 1, selecione os dados referentes à "Energia" do índice "Nível 1 - Setor".

emissao_energia = emissao_por_ano.groupby(['Estado', 'Nível 1 - Setor'])[['Emissão']].sum()
```

```python
emissao_energia.groupby('Nível 1 - Setor').get_group('Energia') # Na primeira vez que fiz eu pensei em fazer com metodo groupby & get_group...
```

```python
emissao_energia.xs('Energia', level= 1)    # Ao corrigir, lembrei que também é possivel fazer com método xs() ...
```

```python
#3) Utilizando a tabela construída na atividade 1, encontre a atividade econômica com valor máximo de emissão do Estado de Minas Gerais

emissão_minas_max = emissao_por_ano.groupby(['Estado', 'Nível 1 - Setor'])[['Emissão']].sum()
```

```python
emissão_minas_max.xs('MG', level = 0).idxmax()
```

```python
#4) Obtenha uma tabela contendo a atividade econômica com máxima emissão para cada Estado.

emissao_por_ano.groupby(['Estado', 'Nível 1 - Setor'])[['Emissão']].sum().idxmax()
```

```python
emissao_max_por_estado = emissao_por_ano.groupby(['Estado', 'Nível 1 - Setor'])[['Emissão']].sum()     # minha solução
emissao_max_por_estado = emissao_max_por_estado.groupby('Estado')[['Emissão']].idxmax()
emissao_max_por_estado
```

```python
emissao_max_por_estado.groupby(level=0).idxmax()   # solução do professor
```

```python
# testando se minha solução tem o mesmo resultado da solução do professor...

Solução_professor = emissao_max_por_estado.groupby(level=0).idxmax()
Minha_solução  = emissao_max_por_estado.groupby('Estado')[['Emissão']].idxmax()

Minha_solução == Solução_professor
```

```python
#5) Obtenha uma tabela contendo o Estado com máxima emissão para cada atividade econômica.

emissao_por_setor_estado = emissao_por_ano.groupby(['Estado', 'Nível 1 - Setor'])[['Emissão']].sum()
emissao_por_setor_estado.groupby('Nível 1 - Setor').idxmax()
```
