## Aula 03 - Transformando os dados

### Desafio 1

```python
#codigo para cotações Yfinance

import yfinance as yf

carteira = ['ASAI3.SA', 'EQTL3.SA', 'LREN3.SA', 'PRIO3.SA', 'RAIL3.SA', 'RDOR3.SA', 'SBSP3.SA', 'VIVT3.SA']

df_cot = yf.download(carteira, start = '2023-01-01', end = '2023-08-31')
df_cot = df_cot.stack(level = 1)

df_cot = df_cot.reset_index().rename(columns = {'Ticker': 'Ativo'})

cotacoes = df_cot[['Date', 'Open','High', 'Low', 'Close', 'Ativo']]

del df_cot, carteira
```

```python
#codigo para indicadores fundamentus

import fundamentus
import pandas as pd

carteira_fun = ['ASAI3', 'EQTL3', 'LREN3', 'PRIO3', 'RAIL3', 'RDOR3', 'SBSP3', 'VIVT3']

df_fun = fundamentus.get_papel(carteira_fun)

df_fun.reset_index(inplace=True)

df_fun =  df_fun[['index','Setor', 'Cotacao', 'Min_52_sem', 'Max_52_sem', 'Nro_Acoes', 'Valor_de_mercado', 'Patrim_Liq', 'Receita_Liquida_12m', 'Receita_Liquida_3m', 'Lucro_Liquido_12m', 'Lucro_Liquido_3m']]
df_fun.rename(columns = {'index':'Ativo'}, inplace= True)

df_fun2 = fundamentus.get_resultado_raw()
df_fun2.reset_index(inplace=True)

df_fun2 = df_fun2.query('papel in @carteira_fun')
df_fun2 = df_fun2.rename(columns= {'papel':'Ativo'})

df_fun2 = df_fun2[['Ativo','P/L','P/VP','Div.Yield','ROE']]
df_fun2.rename(columns={'Div.Yield':'DY'}, inplace=True)

indicadores  = pd.merge(df_fun,df_fun2, on = 'Ativo')

colunas = ['Cotacao', 'Min_52_sem', 'Max_52_sem', 'Nro_Acoes',
       'Valor_de_mercado', 'Patrim_Liq', 'Receita_Liquida_12m',
       'Receita_Liquida_3m', 'Lucro_Liquido_12m', 'Lucro_Liquido_3m']

indicadores[colunas] = indicadores[colunas].apply(pd.to_numeric, errors = 'coerce', axis = 1)

indicadores['LPA'] = indicadores['Lucro_Liquido_12m'] / indicadores['Nro_Acoes']
indicadores['VPA'] = indicadores['Patrim_Liq'] / indicadores['Nro_Acoes']

del carteira_fun, df_fun, df_fun2, colunas
```
