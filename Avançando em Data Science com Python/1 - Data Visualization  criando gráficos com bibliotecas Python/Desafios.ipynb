````python
import pandas as pd
````

````python
url = 'https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/main/Dados/imigrantes_canada.csv'
````

````python
df = pd.read_csv(url)
````

````python
df
````

````python
df.set_index('País', inplace=True)
df
````

````python
anos = list(map(str, range(1980, 2014)))
brasil = df.loc['Brasil', anos]
argentina = df.loc['Argentina', anos]
brasil, argentina
````

````python
brasil_dict = {"ano":brasil.index.tolist(),"Brasil":brasil.values.tolist(), 'Argentina':argentina.values.tolist()}
````

````python
dados_grafico = pd.DataFrame(brasil_dict)
dados_grafico
````

````python
import matplotlib.pyplot as plt
````

````python
plt.plot(dados_grafico['ano'], dados_grafico['Brasil'], label='Brasil')
plt.plot(dados_grafico['ano'], dados_grafico['Argentina'], label='Argentina')
plt.xlabel('Ano')
plt.ylabel('Número de Imigração')
plt.legend()
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.title('Imigração do Brasil e do Canadá de 1980 a 2013')
plt.show()
````

Com este gráfico, podemos notar que as tendências de imigração no Brasil e na Argentina eram bastante semelhantes no início do período. No entanto, a partir dos anos 2000, os números de imigrantes na Argentina começaram a aumentar e atingiram os valores mais elevados entre 2000 e 2005, antes de cair e atingir valores muito baixos nos anos mais recentes. Por outro lado, no caso do Brasil, os aumentos na imigração começaram após 2005 e parecem ter atingido o valor mais alto em 2010.

````python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
````

````python
df = pd.DataFrame(vendas_2022, index= lojas)
df
````

````python
import matplotlib.pyplot as plt
````

````python
fig, axs = plt.subplots(2,2, figsize=(22,8))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Vendas das logas ao longo do ano')

axs[0,0].plot(df.loc['A'])
axs[0,0].set_title('Loja A')

axs[0,1].plot(df.loc['B'])
axs[0,1].set_title('Loja B')

axs[1,0].plot(df.loc['C'])
axs[1,0].set_title('Loja C')

axs[1,1].plot(df.loc['D'])
axs[1,1].set_title('Loja D')

for ax in axs.flat:
  ax.set_ylabel('Vendas')
  ax.set_xlabel('Mês')

ymin = 0
ymax = 500

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)
````

Dica: para escolher a paleta de cores, você também pode consultar a documentação da biblioteca Matploltib. Para aplicar uma paleta de cores a todas as linhas da figura você pode usar a função sns.set_palette() e passar a ela o nome da paleta escolhida.

Estamos empolgados para ver o resultado do seu trabalho e as histórias que você irá contar através deste gráfico. Mãos à obra e divirta-se!

````python
url = 'https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/main/Dados/imigrantes_canada.csv'
df = pd.read_csv(url)
````

````python
df
````

````python
df.set_index('País', inplace=True)
````

````python
anos = list(map(str, range(1980, 2014)))
````

````python
import seaborn as sns
````

````python
sns.set_theme()
sns.set_palette('Dark2')

fig, ax = plt.subplots(figsize=(10, 5))
ax = sns.lineplot(df.loc['Brasil', anos], label='Brasil', lw=3)
ax = sns.lineplot(df.loc['Argentina', anos], label='Argentina', lw=3)
ax = sns.lineplot(df.loc['Peru', anos], label='Peru', lw=3)
ax = sns.lineplot(df.loc['Colômbia', anos], label='Colômbia', lw=3)

ax.set_title('Imigração dos maiores países da América do Sul\npara o Canadá de 1980 a 2013', loc='left', fontsize=20)
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('Número de imigrantes', fontsize=14)

ax.xaxis.set_major_locator(plt.MultipleLocator(5))

ax.legend(title='Países', loc='upper right', bbox_to_anchor=(1.18, 1.02))

plt.show()
````
